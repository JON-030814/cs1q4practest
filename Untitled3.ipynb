```python
def evaluate_numbers(count):
  total = 0
  if count == 0:
    print("No numbers to evaluate.")
    return

  print(f"Please Enter {count} numbers:")
  for i in range(count):
    while True:
      try:
        num = float(input(f"Enter number {i + 1}: "))
        # Check for non-negative numbers as implied by original message
        if num < 0:
          print("Invalid input: Please enter a non-negative number.")
          continue # Ask for input again
        break # Exit while loop if input is valid and non-negative
      except ValueError: # Corrected syntax for exception handling
        print("Invalid input: Please enter a valid number.")

    if num % 2 == 0: # Used 'num' instead of 'number'
      print(f"The number {num} is Even.")
    else:
      print(f"The number {num} is Odd.")
    total += num # Used 'num'

  # Calculate average only if count is greater than 0 to avoid ZeroDivisionError
  if count > 0:
    average = total / count
    print(f"\nTotal of the numbers: {total}")
    print(f"Average of the numbers: {average:.2f}")
  else:
    print("Cannot calculate average for zero numbers.")

# Call the function with a hardcoded number for demonstration.
# The original call `evaluate_numbers(num)` would lead to a NameError
# as 'num' is a local variable inside the function and not defined globally.
evaluate_numbers(3)
```
